```python
continents=visa_dataset['continent'].unique()
certified=[]
denied=[]

for i in continents:
    cond1=visa_dataset['continent']==i
    cond2=visa_dataset['case_status']=='Certified'
    cond3=visa_dataset['case_status']=='Denied'
    count_certified=len(visa_dataset[cond1&cond2])
    count_denied=len(visa_dataset[cond1&cond3])    
    certified.append(count_certified)
    denied.append(count_denied)
```

```python
pd.DataFrame(zip(certified,denied),columns=['Certified','Denied'],index=visa_dataset['continent'].unique())
```
